```python
# import pandas Library , it used for handling datasets , tables
# pandas named as pd to calling short name , so everytime we no need to write full name
import pandas as pd
```

```python
# from pandas library using read_csv() function to read the file and assign to dataset variable
dataset = pd.read_csv("Salary_Data.csv")
dataset
```

```python
# from dataset input column assign to independent variable
independent =dataset[["YearsExperience"]] # if we want more column the dataset[["column1","column2","column3"]]
independent
```

```python
# from dataset output column assign to dependent variable
dependent =dataset[["Salary"]]
dependent
```

```python
# seperate train_set & test_set , in machine learing here we use sklearn library to seperate train_set & test_set
# from sklearn file model-selection library import train_test_split function
from sklearn.model_selection import train_test_split
# this function is parameterized (input,output,test_size=0.30,random_state=0),test size we use 1/3 of dataset
# train_test_split() will give us 4 type variable name as x_train , x_test,y_train,y_test
#for function outputs we assign variables
x_train,x_test,y_train,y_test = train_test_split(independent,dependent,test_size=0.30,random_state=0) 
```

```python
# learning phase 
# from sklearn file linear_model library import LinearRegression function(non parameterized)
from sklearn.linear_model import LinearRegression
# assign variable to LinearRegression function , so we can use dot operator after that variable to call that function
regressor = LinearRegression() 
# calling fit() function from LinearRegression and pass x-train & y_train data ,once excute fit function model created
regressor.fit(x_train,y_train)
```

```python
# y=wX+b , here weight(w) is mentioned in linearRegression function as coef_ (coefficientஒரு பொருளின் குறிப்பிட்ட பண்பை அளவிடும் ஓர் எண்)
# linearRegression formula w is saved in linearRegression function under coef_ name(variable),weight= slop, slop growing ratio
weight = regressor.coef_
weight
```

```python
# linearRegression formula bais(b) is saved in linearRegression function under intercept_ name(variable), b =start poit,min value,intercept
bais = regressor.intercept_
bais
```

```python
# Model saved inside this regressor variable (weight & bais)
# next is evaluating the model
# predict() function also inside LR , we assign variable y_pred and input the x_test variable(test data test)
# here we not require y_test data because this actual data for x-test , we are going to see how the model is predicting based on x_test data
y_pred = regressor.predict(x_test) # AI Brain giving output(predicting)
y_pred
```

```python
# next we validating our model by finding R^2 value
# importing r2_score function from sklearn
from sklearn.metrics import r2_score
# assign variable to r2_score function , input parameter is y_test(actual value) and y_pred (predicted value)
# here we compare actual value with predicted value to find the result how close to actual value , near to 1 is good result
# r2 values(result) are between 0 to 1 only
r_score = r2_score(y_test,y_pred)
```

```python
r_score
```

```python
# once you feel your model is working good next step is saving the model
# here we using pickle module to save the model
# pickle filesave extension is .sav
import pickle
# give file name with algorithm name ( easy to understant what algorithm used),
filename = "finalized_model_linear.sav" # file name only created
```

```python
# dumping regressor model into the finalized_model_linear file
# "wb" is write binary and "rb" is read binary
pickle.dump(regressor,open(filename,"wb"))
```

```python
# next step is end user/ deployment
# open the file in read mode assign to loaded_model variable
loaded_model = pickle.load(open("finalized_model_linear.sav","rb"))
# calling the predict function from loaded model
result = loaded_model.predict([[13]])
```
